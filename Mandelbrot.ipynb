```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit

%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
@jit
def frac_funk(c):
    z = c
    for i in range(60):
        z1 = z**2 + c
        if np.abs(z) > 2:
            return i
        z=z1
    return 60

@jit
def zpoints(xs, ys):
    S = np.zeros((xs.size,ys.size))
    for i,x in enumerate(xs):
        for w,y in enumerate(ys):
            z = y + x*1j
            S[i,w] = frac_funk(z)
    return S
```

```python
a = np.linspace(-1.5,1.5,10000)
b = np.linspace(-2,1,10000)

%time sol = zpoints(a,b)
```

```python
plt.matshow(sol, cmap='YlGnBu')
plt.axis('off')
```
